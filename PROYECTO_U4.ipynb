## INTRODUCCIÓN

La educación es un pilar fundamental en el desarrollo de las sociedades, y su calidad depende de una amplia gama de factores. Este proyecto analiza cómo diferentes variables afectan el desempeño académico de los estudiantes, específicamente su puntaje en exámenes estandarizados. Variables como la participación parental, acceso a recursos educativos, motivación, actividades extracurriculares, calidad del profesorado y características socioeconómicas, entre otras, se estudian a través de análisis estadísticos. Al emplear herramientas como el ANOVA y la prueba de Tukey, se busca identificar patrones significativos y relaciones entre estas variables. Los resultados obtenidos pueden orientar estrategias educativas y políticas públicas que promuevan una educación más equitativa y efectiva, beneficiando tanto a estudiantes como a las comunidades escolares.

## MARCO TEÓRICO

El rendimiento académico de los estudiantes es un fenómeno complejo influenciado por factores individuales, familiares y contextuales. En el ámbito educativo, se ha comprobado que el entorno familiar, el acceso a recursos y la motivación son determinantes clave. Según el modelo ecológico de Bronfenbrenner, las interacciones entre el entorno inmediato (familia, escuela, compañeros) y los recursos externos (infraestructura, tecnología) moldean el desarrollo del estudiante.

El uso de análisis estadísticos como el ANOVA permite identificar diferencias significativas en grupos independientes, mientras que la prueba de Tukey profundiza en estas diferencias para evaluar comparaciones específicas entre niveles de las variables. Por ejemplo, investigaciones previas han encontrado que la participación parental tiene un impacto positivo directo en el rendimiento académico, ya que fomenta hábitos de estudio, apoyo emocional y motivación. Este proyecto se centra en una perspectiva analítica y empírica para validar estas relaciones en un contexto específico.

# **PARTICIPACIÓN** **PARENTAL**

```python
df['participacion_parental'].unique()
```

```python
import scipy.stats as stats
nivel_de_significancia =0.05

#Realizar ANOVA
anova_result = stats.f_oneway(
      df[df['participacion_parental'] == 'bajo']['puntaje_examen'],
      df[df['participacion_parental'] == 'medio']['puntaje_examen'],
      df[df['participacion_parental'] == 'elevado']['puntaje_examen'],

)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadístico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusión: Hay diferencias significativas entre al menos dos de los tratamientos.")
else:
    print("Conclusión: No hay diferencias significativas entre los tratamientos.")
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['participacion_parental'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Gráfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")
```

# CONCLUSIÓN

El análisis mostró que la participación de los padres tiene un efecto significativo en el puntaje de los estudiantes. Aquellos con padres que participan activamente en su educación, ya sea supervisando tareas o asistiendo a reuniones escolares, tienden a obtener mejores resultados. Este hallazgo respalda investigaciones previas que enfatizan el papel crucial de la familia en el desarrollo académico

Variables como la motivación, el acceso a recursos, la calidad del profesorado y el entorno familiar tienen un impacto significativo y directo en los resultados educativos. Estas conclusiones destacan la necesidad de políticas educativas integrales que aborden tanto las necesidades individuales como las colectivas, promoviendo una educación más inclusiva y equitativa. Asimismo, se sugiere continuar investigando y aplicando estos hallazgos en distintos contextos para fortalecer el sistema educativo en general.
